```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('matches - matches.csv')
data
```

```python
delivery=pd.read_csv('deliveries.csv')
delivery
```

```python
new_data=delivery.merge(data,left_on='match_id',right_on='id')
new_data
```

```python
new_data.groupby(['season','bowler'])['player_dismissed'].count().sort_values(ascending=False).reset_index().drop_duplicates(subset='season',keep='first').sort_values('season')[['season','bowler']]
```

```python
top_batsman=new_data.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
top_batsman
```

```python
top_bowler=new_data.groupby('bowler')['player_dismissed'].count().sort_values(ascending=False).head(10)
top_bowler
```

```python
batsman_list=top_batsman.index.tolist()
final=delivery[delivery['batsman'].isin(batsman_list)] #includes the data of top 10 batsman only
bowler_list=top_bowler.index.tolist()
final2=final[final['bowler'].isin(bowler_list)]
final2  #includes the data of top 10 batsman and top 10 bowlers only
```

```python
pt=final2.pivot_table(index='bowler',columns='batsman',values='batsman_runs',aggfunc='sum')
pt
```

```python
import seaborn as sb
sb.heatmap(pt)
```
