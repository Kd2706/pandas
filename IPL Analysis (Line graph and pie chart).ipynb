```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('matches - matches.csv')
data
```

```python
delivery=pd.read_csv('deliveries.csv')
delivery
```

```python
new_data=delivery.merge(data,left_on='match_id',right_on='id')
new_data
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
def batsman(name):
    
    mask=new_data[new_data['batsman']==name]
    mask2=mask.groupby('season')['batsman_runs'].sum()
    plt.plot(mask2, marker='D', markersize=5)
    plt.xlabel("season")
    plt.ylabel("Batsman runs")
    plt.title("Season by season batsman analysis")
    plt.show()

name=input("Enter the batsman name whose run is to be analysed: ")
batsman(name)
```

```python
def batsman(name):
    
    mask=new_data[new_data['batsman']==name]
    runs6=mask[mask['batsman_runs']==6]['batsman_runs'].sum()
    runs4=mask[mask['batsman_runs']==4]['batsman_runs'].sum()
    runs3=mask[mask['batsman_runs']==3]['batsman_runs'].sum()
    runs2=mask[mask['batsman_runs']==2]['batsman_runs'].sum()
    runs1=mask[mask['batsman_runs']==1]['batsman_runs'].sum()
    areas=["6's","4's","3's","2's","1's"]
    runs=[runs6,runs4,runs3,runs2,runs1]
    plt.pie(runs, labels=areas, autopct='%1d%%')
    plt.show()

name=input("Enter the batsman name whose run is to be analysed: ")
batsman(name)
```
